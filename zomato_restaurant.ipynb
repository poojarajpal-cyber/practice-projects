```python
#lets split the dataset into features and label



x=data.drop(columns=['Average Cost for two'])
y=data['Average Cost for two']
```

```python
# lets import necessary libraries
# lets train_test and split

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,r2_score,classification_report
from sklearn.neighbors import KNeighborsRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
lr=LinearRegression()
```

```python
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=i)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    print(f"at random state{i}the training accuracy is:{r2_score(y_train,pred_train)}")
    print(f"at random state{i}the testing  accuracy is:{r2_score(y_test,pred_test)}")
    print("\n")
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=70)
```

```python
lr.fit(x_train,y_train)
```

```python
pred_test=lr.predict(x_test)
```

```python
print(r2_score(y_test,pred_test))
```

```python
# cross validation 
train_accuracy=r2_score(y_train,pred_train)
test_accuracy=r2_score(y_test,pred_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
for i in range(2,10):
    cv_score=cross_val_score(lr,x,y,cv=i)
    cv_mean=cv_score.mean()
    print(f"at cross fold{i}the cv score is{cv_mean}and accuracy score for training is{train_accuracy}and accuracy score for testing is{test_accuracy}")
    print("\n")
```

```python
#regularization
from sklearn.linear_model import Lasso
```

```python
parameters={'alpha':[.0001,.001,.01,.1,1],
           'random_state':list(range(0,10))}
ls=Lasso()
clf=GridSearchCV(ls,parameters)
clf.fit(x_train,y_train)
print(clf.best_params_)
```

```python
# final model training
ls=Lasso(alpha=0.001,random_state=0)
ls.fit(x_train,y_train)
ls_score=ls.score(x_train,y_train)
pred_ls=ls.predict(x_test)
ls_score*100
lss=r2_score(y_test,pred_ls)
lss*100
```
