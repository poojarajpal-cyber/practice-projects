```python
new_df=df.copy()
```

```python
# as we can see that there is a strong correlation between months as a customer and age 
# and between vehicle claim and total claim amount 
# lets delete any one column
df=new_df.drop(['vehicle_claim','months_as_customer'],axis=1,inplace=True)
```

```python
# lets divide the dataset into label and features


x=new_df.drop(columns=['fraud_reported'],axis=1)
y=new_df['fraud_reported']
```

```python
# lets train_test and split
# for that importing necessary libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,r2_score,classification_report
from sklearn.neighbors import KNeighborsClassifier
```

```python
#lets train_test and split the dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=40)
```

```python
def metric_score(clf,x_train,x_test,y_train,y_test,train=True):
    if train:
        y_pred=clf.predict(x_train)
        print("\n=======train result========")
        print(f"accuracy score:{accuracy_score(y_train,y_pred)*100:.2f}%")
    elif train==False:
        pred=clf.predict(x_test)
        print("\n===============test result===========")
        print(f"accuracy score:{accuracy_score(y_test,pred)*100:.2f}%")
        
```

```python
# initializing model
knn=KNeighborsClassifier()
```

```python
# model training
knn.fit(x_train,y_train)
```

```python
#calling the function and pass the dataset and checking the score
metric_score(knn,x_train,x_test,y_train,y_test,train=True)
metric_score(knn,x_train,x_test,y_train,y_test,train=False)
```

```python
# cross validation
from sklearn.model_selection import cross_val_score
```

```python
# cross validation to check if the model is overfitting or not
cross_val_score(knn,x,y,cv=5)
```

```python
cross_val_score(KNeighborsClassifier(),x,y,cv=5).mean()
```

```python
#hyper parameter tuning
param_grid={'algorithm':['kd_tree','brute'],
           'leaf_size':[3,5,6,7,8],
           'n_neighbors':[3,5,7,9,11,13]}
grd_srch=GridSearchCV(estimator=knn,param_grid=param_grid)
```

```python
grd_srch.fit(x_train,y_train)
grd_srch.best_params_
```

```python
knn=KNeighborsClassifier(algorithm='kd_tree',leaf_size=3,n_neighbors=13)
knn.fit(x_train,y_train)
```

```python
#calling the function and pass the dataset and checking the score
metric_score(knn,x_train,x_test,y_train,y_test,train=True)
metric_score(knn,x_train,x_test,y_train,y_test,train=False)
```

```python
# lets try with another model
clf=DecisionTreeClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
#calling the function and pass the dataset and checking the score
metric_score(clf,x_train,x_test,y_train,y_test,train=True)
metric_score(clf,x_train,x_test,y_train,y_test,train=False)
```

```python
# hyper parameter tuning
grid_param={'criterion':['gini','entropy'],
           'max_depth':range(10,15),
           'min_samples_leaf':range(2,6),
           'min_samples_split':range(3,8),
           'max_leaf_nodes':range(5,10)}
```

```python
grd_srch=GridSearchCV(estimator=clf,param_grid=grid_param,cv=5)
grd_srch.fit(x_train,y_train)
```

```python
best_parameters=grd_srch.best_params_
best_parameters
```

```python
clf=DecisionTreeClassifier(criterion='entropy',max_depth=10,max_leaf_nodes=5,min_samples_leaf=2,min_samples_split=3)
```

```python
clf.fit(x_train,y_train)
```

```python
#calling the function and pass the dataset and checking the score
metric_score(clf,x_train,x_test,y_train,y_test,train=True)
metric_score(clf,x_train,x_test,y_train,y_test,train=False)
```
