```python
# we will encode our region column using label encoder
from sklearn.preprocessing import LabelEncoder
```

```python
#initialising
lab_enc=LabelEncoder()
data2=lab_enc.fit_transform(data['Region'])
pd.Series(data2)
data['Region']=data2
data.head()
```

```python
data.drop(columns=['Country'],axis=1,inplace=True)
```

```python
#there seems no such skewness in the dataset
# as we have seen there is strong coreelation between two columns namely so we delete one column
x=data.drop(columns=['Health (Life Expectancy)','Happiness Score'],axis=1)
y=data['Happiness Score']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=40)
```

```python
data.info()
```

```python
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python
#lets check how well model fits the train data
regression.score(x_train,y_train)
```

```python
#lets check how well the model fits the test data
regression.score(x_test,y_test)
```
