```python
# lets train_test and split
# for that importing necessary libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,r2_score,classification_report
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVC
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')
```

```python
# lets divide the dataset into label and features


x=df_1.drop(columns=['RainTomorrow'],axis=1)
y=df_1['RainTomorrow']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=355)
```

```python
# training the model
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=100,random_state=0)
rf.fit(x_train,y_train)
```

```python
def metric_score(clf,x_train,x_test,y_train,y_test,train=True):
    if train:
        y_pred=clf.predict(x_train)
        print("\n=======train result========")
        print(f"accuracy score:{accuracy_score(y_train,y_pred)*100:.2f}%")
    elif train==False:
        pred=clf.predict(x_test)
        print("\n===============test result===========")
        print(f"accuracy score:{accuracy_score(y_test,pred)*100:.2f}%")
        
```

```python
# lets check the score
metric_score(rf,x_train,x_test,y_train,y_test,train=True)
metric_score(rf,x_train,x_test,y_train,y_test,train=False)
```

```python
y_test=y_test.values.reshape(-1,1)
print(y_test)
```

```python
y_pred=(rf.predict(x_test))
```

```python
print(y_pred)
```

```python
#hyper parameter tuning
params={'n_estimators':[13,15],
       'criterion':['entropy','gini'],
       'max_depth':[10,15],
       'min_samples_split':[10,11],
       'min_samples_leaf':[5,6]}
grd=GridSearchCV(rf,param_grid=params)
grd.fit(x_train,y_train)
print(grd.best_params_)
```

```python
rf=grd.best_estimator_
rf.fit(x_train,y_train)
```

```python
# lets check the score
metric_score(rf,x_train,x_test,y_train,y_test,train=True)
metric_score(rf,x_train,x_test,y_train,y_test,train=False)
```

```python
# check the ccuracy using another model
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
# lets check the score
metric_score(dt,x_train,x_test,y_train,y_test,train=True)
metric_score(dt,x_train,x_test,y_train,y_test,train=False)
```
